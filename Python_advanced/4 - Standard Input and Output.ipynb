# Output Formatting

```python
a = 10; b = 20 #multiple statements in single line.

print("The value of a is {} and b is {}".format(a, b))    #default
```

1. Syntax & Readability
Style	Example
% formatting	"%s is %d years old" % (name, age)
.format()	"{0} is {1} years old".format(name, age)

With %, you use format specifiers (%s, %d, %f, etc.) inside the string, and provide a tuple (or single value) after the %.

With .format(), you embed {} placeholders and call .format(...) with positional or named arguments.

2. Positional vs. Named Arguments
% formatting supports only positional substitution:


"%s scored %d points" % ("Alice", 42)
.format() lets you use both positional and named fields:


"{player} scored {points} points".format(player="Alice", points=42)

"{0} scored {points} points".format("Alice", points=42)
3. Reuse & Reordering
With %, you cant easily reuse the same value twice without repeating it in the tuple:


"%s %s %s" % ("spam", "eggs", "ham")  # each appears once only
With .format(), you can refer to the same argument multiple times, or reorder without changing the call:


"{0}, {1}, and {0} again".format("spam", "eggs")

4. Advanced Formatting Options
Alignment, width, padding, precision

% supports minimal width/precision:


"%10s" % "foo"   # right-align in width 10
"%.2f" % 3.14159 # 2 decimal places
.format() has a richer mini-language inside {}:


"{:>10}".format("foo")     # right-align width 10
"{:.2f}".format(3.14159)   # 2 decimal places
"{:^10}".format("spam")    # centered in width 10
"{:0>5d}".format(42)       # pad number with zeros → "00042"
5. Automatic Type Conversion & Safety
%s will convert any object via str(), but other specifiers (%d, %f) require matching types (or you get a TypeError).

.format() is generally more forgiving: if you use {} without a format spec, it calls str() on the object.

6. Nesting & Object Attribute Access
.format() can drill into attributes or items:


class P: pass
p = P(); p.x = 42
"Value: {0.x}".format(p)        # → "Value: 42"
"Dict value: {d[key]}".format(d={"key": "val"})
% has no built-in support for attribute/item lookup—you’d have to pre-extract values.

7. Deprecation & Modern Alternatives
The % operator is considered legacy; for new code, .format() or even better, f-strings (Python 3.6+) are preferred:


f"{name} scored {points:.2f} points"

```python
a = 10; b = 20  #multiple statements in single line

print("The value of b is {1} and a is {0}".format(a, b)) #specify position of arguments
```

```python
#we can use keyword arguments to format the string
print("Hello {name}, {greeting}".format(name="rahul", greeting="Good Morning"))
```

```python
#we can combine positional arguments with keyword arguments
print('The story of {0}, {1}, and {other}'.format('Bill', 'Manfred',
                                                       other='Georg'))
```
