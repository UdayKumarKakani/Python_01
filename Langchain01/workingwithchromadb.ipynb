```python
from langchain_together import ChatTogether,TogetherEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from pypdf import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
filepath = r"D:\Python_01\Python_01\Langchain01\Introduction_to_Data_and_Data_Science.pdf"  # Replace with your PDF file path
```

```python
reader = PdfReader(filepath)
text = ""
for page in reader.pages:
    text += page.extract_text() 
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)
chunks=splitter.split_text(text)
print(chunks[0])  # Print first chunk for verification
```

```python
embedding = TogetherEmbeddings(
        model="BAAI/bge-base-en-v1.5",
        api_key=os.getenv("TOGETHER_API_KEY")
    )
pd = "./intro-cromedb"
    # Initialize Chroma vectorstore (empty)
vectorstore = Chroma(
        embedding_function=embedding,
        persist_directory=pd
    )
```

```python
for i, chunk in enumerate(chunks):
        document = Document(embedding=embedding,page_content=chunk, metadata={"source": filepath, "chunk_index": i})
        try:
            id=vectorstore.add_documents([document])
            print(id)
            print(f"Stored chunk {i+1}/{len(chunks)}")
        except Exception as e:
            print(f"Failed to store chunk {i+1}: {e}")
vectorstore.persist()
```

```python
# Get all documents from the vectorstore
all_docs = vectorstore.get(include=["documents", "metadatas", "embeddings"])

# Access the first document if it exists
if all_docs["documents"]:
    first_doc = all_docs["documents"][0]
    print("First document content:")
    print(first_doc)
    # If you want its metadata:
    print("Metadata:")
    print(all_docs["metadatas"][0])
    # If you want its embedding:
    print("embeddings:")
    print(all_docs["embeddings"][0])
else:
    print("No documents found in Chroma DB.")
```
